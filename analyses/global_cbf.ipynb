```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import statsmodels.formula.api as smf
%matplotlib inline

repos_dir="/media/veracrypt1/repos/ASL_longitudinal/analyses/"

df=pd.read_csv(repos_dir+"graphs.csv", sep=',')

plt.rcParams["figure.figsize"] = [8, 5]
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'

x1 = df["Age"][df["Male"]==1]
y1 = df["CBF"][df["Male"]==1]
plt.plot(x1,y1,'o-', color=sns.xkcd_rgb["denim blue"], linewidth=.3)

x2=df["Age"][df["Male"]==0]
y2=df["CBF"][df["Male"]==0]
plt.plot(x2,y2,'*-', color=sns.xkcd_rgb["pale red"], linewidth=.3)

plt.plot(np.unique(x1), np.poly1d(np.polyfit(x1, y1, 1))(np.unique(x1)), '-', color=sns.xkcd_rgb["denim blue"], linewidth=2)
plt.plot(np.unique(x2), np.poly1d(np.polyfit(x2, y2, 1))(np.unique(x2)), '-', color=sns.xkcd_rgb["pale red"], linewidth=2)

# plt.title("Increase of global CBF with age", fontsize=20)
plt.xlabel('Age (years)')
plt.ylabel('CBF (mL/100g/min)')
plt.ylim(20,130)

# plt.show()

plt.savefig('cbf_global.png', dpi=300)
```
